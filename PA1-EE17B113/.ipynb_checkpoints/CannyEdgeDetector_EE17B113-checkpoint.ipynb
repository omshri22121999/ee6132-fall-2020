```python
# Importing the required libraries

import numpy as np
import scipy.stats as st
import scipy.signal as sg
import matplotlib.pyplot as plt
import cv2

# To remove unnecessary warnings
import warnings

warnings.filterwarnings("ignore")

# Note : Used for autoformatting the notebook, remove if causing any error
%load_ext nb_black
```

## Canny Edge Detector

**The steps involved in this implementation of Canny Edge Detector is as follows :**
1. First we read the image and covert it to black and white
2. We apply Gaussian Smoothing on the image
3. We next apply the Sobel filters on the image to get the horizontal and vertical derivatives using which we will calculate the magnitude and direction of gradient.
4. We then apply Non-Max Supression on the image to thin the edges.
5. We apply single thresholding to remove weak pixels and set strong pixels to the highest value to get the edges.

### Step 1:

**Loading the image**

```python
# Loading the image

img = cv2.imread("./clown.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.title("Original Image")
plt.axis("off")
plt.imshow(img)
plt.show()
```

**Converting the image to grayscale**

```python
# Converting the image to grayscale

img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.title("Grayscale Image")
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.show()
```

### Step 2:

**Apply Gaussian Filter of Kernel Size = 5x5 and Sigma = 1.5** 

```python
# Function to apply Gaussian Filter


def GaussianFilter(img, ker_size, sig):

    # Creating an 1d Space
    x = np.linspace(-2 / sig, 2 / sig, ker_size)

    # Generating 1d Gaussian Kernel by taking the difference of CDF of the above space
    gauss_1d = st.norm.pdf(x)

    # Creating the 2d Gaussian Kernel by multiplying two 1d Gaussian Kernels
    gauss_2d = np.outer(gauss_1d, gauss_1d)

    # Normalizing the Gaussian Kernel
    gauss_kernel = gauss_2d / gauss_2d.sum()

    # Convolving the filter
    img_fil = sg.convolve2d(img, gauss_kernel, mode="same", boundary="symm")

    return img_fil
```

```python
# Applying the Gaussian Filter to the grayscale image :

img_fil = GaussianFilter(img_gray, 5, 1.5)
```

```python
plt.title("Grayscale Image Gaussian Filtered")
plt.axis("off")
plt.imshow(img_fil, cmap="gray")
plt.show()
```

### Step 3

**Applying Sobel Filters and Extracting the Magnitude and Direction**

```python
# Function to apply Sobel Filter


def SobelFilter(img):

    # Creating Sobel filter

    kern_h = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=np.float)
    kern_v = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], dtype=np.float)

    # Finding derivative using Sobel filters

    img_gx = sg.convolve2d(img, kern_h, mode="same", boundary="symm")
    img_gy = sg.convolve2d(img, kern_v, mode="same", boundary="symm")

    return img_gx, img_gy
```

```python
# Applying the Sobel Filter

img_gx, img_gy = SobelFilter(img_fil)
```

```python
plt.title("Gx")
plt.imshow(img_gx, cmap="gray")
plt.axis("off")
plt.show()

plt.title("Gy")
plt.imshow(img_gy, cmap="gray")
plt.axis("off")
plt.show()
```

**Extract Magnitude and Direction**

```python
# Extracting Magnitude
img_mag = np.hypot(img_gx, img_gy)

# Normalizing the Magnitude to Pixel Intensity Range (0-255)
img_mag *= 255 / img_mag.max()

# Extracting Angle
img_ang = np.arctan2(img_gy, img_gx) * 180 / np.pi

# Converting negative angles to corresponding positive angle for direction
img_ang[img_ang < 0] += 180
```

```python
plt.title("Magnitude")
plt.imshow(img_mag, cmap="gray")
plt.axis("off")
plt.show()

# Displaying Direction using HSV Representation (H => Angle and V => Magnitude (Saturation = 255)

img_ang_disp = np.dstack(
    [
        np.round(img_ang),
        255 * np.ones(img_ang.shape, dtype=np.uint8),
        np.round(img_mag),
    ]
).astype(np.uint8)

img_ang_disp = cv2.cvtColor(img_ang_disp, cv2.COLOR_HSV2RGB)

plt.title("Direction")
plt.imshow(img_ang_disp)
plt.axis("off")
plt.show()
```

### Step 4

**Algorithm for Applying Non-Max Supression**

- To simplify the algorithm, we use the nearest neighbours as already existing pixels in image instead of interpolating the pixels.

- The range of angles and respective pixels are as follows :
    - $0^o-22.5^o$ (or) $157.5^o-180^o$ : We consider the pixels to the *left* and *right* of the given pixel as neighbours.
    - $22.5^o-67.5^o$ : We consider the pixels at the *top-right* and *bottom-left* of the given pixel as neighbours.
    - $67.5^o - 112.5^o$ : We consider the pixels at the *top* and *botton* of the given pixel as the neigbours.
    - $112.5^o - 157.5^o$ : We consider the pixels at the *bottom-right* and *top-left* of the given pixel as the neighbours.
    
- We iterate through each pixel and based on the above criteria, we select two neighbours for the pixel

- If the pixel magnitude is the greater than its neighbour, we assume the pixel to be a maximum and copy it's value in the pixel's location in placeholder. Otherwise, we set the magnitude value as 0 in the pixel's location in the placeholder.

```python
def NonMaxSuppr(img_mag, img_ang):

    # Placeholder for Non-Max Supression Output
    img_mag_max = np.zeros(img_mag.shape)

    # Applying above alogorithm

    for i in range(img_mag.shape[0]):
        for j in range(img_mag.shape[1]):

            ## Placeholder for neighbours

            p = 255
            q = 255

            ## Assigning the pixel magnitude value

            ### 0-22.5 / 157.5-180 => LEFT,RIGHT
            if (0 <= img_ang[i, j] < 22.5) or (157.5 <= img_ang[i, j] <= 180):

                if j > 0:
                    p = img_mag[i, j - 1]

                if j < img_ang.shape[1] - 1:
                    q = img_mag[i, j + 1]

            ### 22.5-67.5 => TOP-RIGHT,BOTTOM-LEFT
            elif 22.5 <= img_ang[i, j] < 67.5:

                if i < img_ang.shape[0] - 1 and j > 0:
                    p = img_mag[i + 1, j - 1]

                if i > 0 and j < img_ang.shape[1] - 1:
                    q = img_mag[i - 1, j + 1]

            ### 67.5-112.5 => TOP,BOTTOM
            elif 67.5 <= img_ang[i, j] < 112.5:

                if i < img_ang.shape[0] - 1:
                    p = img_mag[i + 1, j]

                if i > 0:
                    q = img_mag[i - 1, j]

            ### 112.5-157.5 => BOTTOM-RIGHT,TOP-LEFT
            elif 112.5 <= img_ang[i, j] < 157.5:

                if i < img_ang.shape[0] - 1 and j < img_ang.shape[1] - 1:
                    p = img_mag[i + 1, j + 1]

                if i > 0 and j > 0:
                    q = img_mag[i - 1, j - 1]

            ## Comparing the pixel value to the neighbours and assigning pixel value to placeholder
            if img_mag[i, j] >= p and img_mag[i, j] >= q:
                img_mag_max[i, j] = img_mag[i, j]

    return img_mag_max
```

**Applying Non-Max Supression**

```python
# Applying Non-Max Supression
img_mag_max = NonMaxSuppr(img_mag, img_ang)
```

```python
plt.title("Non-Max Supression Output")
plt.axis("off")
plt.imshow(img_mag_max, cmap="gray")
plt.show()
```

### Step 5

**Single Thresholding**

We convert pixel value less than the median of the image magnitude to 0 and those equal to or above it to 255 in the output of Non-Max Supression

```python
# Single Thresholding

median = np.median(img_mag)
img_canny_5_1o5 = img_mag_max.copy()
img_canny_5_1o5[img_mag_max < median] = 0
img_canny_5_1o5[img_mag_max >= median] = 255
```

```python
# Visualizing the Canny Edge Detector Output

plt.title("Canny Kernel Size = 5, Sigma = 1.5")
plt.axis("off")
plt.imshow(img_canny_5_1o5, cmap="gray")
plt.show()
```

Thus, we have applied Canny Edge Detector to extract the edges of the given image

## Comparing output for different _sigma_ values

We have extracted the edges of the image above using _sigma_ value of 1.5, now we will try the same procedure using _sigma_ value of 3.

```python
# Applying the Gaussian Filter to the grayscale image :
img_fil_n = GaussianFilter(img_gray, 5, 3)  # Using kernel size = 5 sigma = 3

# Applying the Sobel Filter
img_gx_n, img_gy_n = SobelFilter(img_fil_n)

# Extracting Magnitude
img_mag_n = np.hypot(img_gx_n, img_gy_n)

# Normalizing the Magnitude to Pixel Intensity Range (0-255)
img_mag_n *= 255 / img_mag_n.max()

# Extracting Angle
img_ang_n = np.arctan2(img_gy_n, img_gx_n) * 180 / np.pi

# Converting negative angles to corresponding positive angle for direction
img_ang_n[img_ang_n < 0] += 180

# Applying Non-Max Supression
img_mag_max_n = NonMaxSuppr(img_mag_n, img_ang_n)

# Single Thresholding
median_n = np.median(img_mag_n)
img_canny_5_3 = img_mag_max_n.copy()
img_canny_5_3[img_mag_max_n < median_n] = 0
img_canny_5_3[img_mag_max_n >= median_n] = 255
```

```python
# Visualizing the Canny Edge Detector Output

plt.title("Canny Kernel Size = 5, Sigma = 3")
plt.axis("off")
plt.imshow(img_canny_5_3, cmap="gray")
plt.show()
```

```python
# Comparing the outputs of sigma = 3 and sigma = 1.5 side-by-side

f, ax = plt.subplots(1, 2)

f.set_figheight(30)
f.set_figwidth(15)

ax[0].title.set_text("Canny Kernel Size = 5, Sigma = 1.5")
ax[0].imshow(img_canny_5_1o5, cmap="gray")
ax[0].axis("off")

ax[1].title.set_text("Canny Kernel Size = 5, Sigma = 3")
ax[1].imshow(img_canny_5_3, cmap="gray")
ax[1].axis("off")

plt.show()
```

We see that by increasing the sigma value, we get lesser edges and the edges becomes slightly more bent which is visible near the text "WHAT'S THE STORY?" and near the nose, due to the slightly excessive blurring.
