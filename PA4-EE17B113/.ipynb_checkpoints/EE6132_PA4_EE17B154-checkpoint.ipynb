### Function to convert image coordinates to the normalized camera coordinates.

```python
def img_to_norm_camera(matched_points, K):
    """
    Using the focal lengths and the principal point coordinates from the camera intrinsic matrix
    convert the  image coordinates to the normalized camera coordinates.
    """
    # Extract intrinsic parameters from the camera matrix
    fx = K[0][0] # Focal length
    fy = K[1][1]
    cx = K[0][2] # Principal point coordinates
    cy = K[1][2]

    normalized_match_points = np.zeros(matched_points.shape)

    # Convert the  image coordinates to the normalized camera coordinates
    normalized_match_points[:, 0, :] = (matched_points[:, 0, :] - cx)/fx
    normalized_match_points[:, 1, :] = (matched_points[:, 1, :] - cy)/fy

    return normalized_match_points
```

### Rotation Function

```python
def RotationFunction(threeDpoints, r):
    """
    r contains the theta vector for one frame.
    This function rotates all the 3D points for that particular frame and the multiplication is written out step-wise
    """

    rotated_threeDpoints = np.zeros(threeDpoints.shape)
    rotated_threeDpoints[:,0] = threeDpoints[:, 0] - threeDpoints[:, 1]*r[2] + threeDpoints[:, 2]*r[1]
    rotated_threeDpoints[:,1] = threeDpoints[:, 0]*r[2] + threeDpoints[:, 1] - threeDpoints[:, 2]*r[0]
    rotated_threeDpoints[:,2] = -1*threeDpoints[:, 0]*r[1] + threeDpoints[:, 1]*r[0] + threeDpoints[:, 2]     

    return rotated_threeDpoints
```

### Projection Function

```python
def ProjectionFunction(threeDpoints, camera_parameters, n_points, n_frames):
    """
    Given the 3D points and all the camera parameters for each frame, this function
    projects the 3D points with respect to each frame
    """
    projected_points = np.zeros((n_frames, n_points, 3))
    
    for i in range(n_frames):
        # i th frame, the first
        R_matrix_angles_i = camera_parameters[i, :3]
        rotated_threeDpoints = RotationFunction(threeDpoints, R_matrix_angles_i)
   
        # To the rotated coordinates, add the translation vector (last 3 values)
        T_matrix_i = camera_parameters[i, 3:]
        projected_points[i] = rotated_threeDpoints + T_matrix_i
        
    # Converting 2D homogeneous to heterogeneous coordinates
    projected_x = projected_points[:, : ,0]/projected_points[:, :, 2]
    projected_y = projected_points[:, :, 1]/projected_points[:, :, 2]
    return projected_x.T, projected_y.T
```

### Reprojection Error Function

```python
def reprojection_error_residuals(params, p, n_frames, n_points):
    """
    Calculates the residuals for reconstructed x and y coordinates
    """
    # Separating the camera rotation and translation parameters and the depth parameters
    camera_parameters = params[: n_frames * 6].reshape((n_frames, 6))
    W = params[n_frames * 6: ].reshape((n_points, 1))

    # Initial 3 D points
    threeDpoints = np.c_[p[:, 0, 0], p[:, 1, 0], np.ones((n_points,1))]/W

    projected_x , projected_y = ProjectionFunction(threeDpoints, camera_parameters, n_points, n_frames)
    
    # Calculate the residuals for x and y coordinates
    errors_in_x = projected_x - p[:, 0, :n_frames]
    errors_in_y = projected_y - p[:, 1, :n_frames]

    return np.hstack((errors_in_x,  errors_in_y)).ravel()
```

### Bundle Adjustment

```python
def bundle_adjustment(n_frames, matched_points):
    """
    Performs bundle adjustments to get estimated of the projection matrices of each frame and the 3D points
    """

    # Normalize the match point correspondences using fx, fy, cx and cy from the intrinsic camera matrix
    p = img_to_norm_camera(matched_points, K) 

    # Number of match point correspondences
    n_points = matched_points.shape[0]

    # Initial values
    x0 = np.ones((2*(n_frames * 3) + n_points))
    x0[:2*(n_frames * 3)] = 0.0

    # Get the estimated parameters using the least squares routine of scipy
    estimated_parameters = scipy.optimize.least_squares(reprojection_error_residuals, x0, args=(p, n_frames, n_points))
    
    np.savetxt("tony_answer.csv",estimated_parameters.x)
    
    return estimated_parameters
```

### Plane Sweep Stereo

1. For a plane at depth $d$ and with normal vector $n$. the plane-induced homography mapping the reference frame to the $i^{th}$ frame is:
$$H_{d, i} = K[R_i - \frac{T_i n^T}{d}]K^{-1}$$

2. We are interested only in planes that are parallel to the reference frame, therefore $n = [0, 0, -1]^T$

3. The matrices $R_i$ (rotation matrix) for each frame, the translation matrix $T_i$ are obtained from bundle adjustment.

4. From the inverse-depth values estimated by bundle adjustment, we first take the reciprocal to get depths. From this, I have used np.linspace to select 10 equally spaces depth values between the minimum and maximum depth.

```python
def plane_sweep_stereo(n_frames, matched_points, images_array):
    """
    Takes the number of frames, match point correspondences and the array of images(frames) as input.
    1. Performs bundle adjustment to get estimates for translation matrix, rotation angles, inverse depth estimates
    2. Chooses 10 candidate planes at equally spaced depth values
    3. Each image is projected to the reference frame using the plane induced homography matrix
    4. This is done for each frame, for every depth value. (we get a h x w x 10 x n_frames matrix)
    5. We then find the variance along the "number of frames" axis.
    6. The value of the depth that gives the least variance is selected as the correct depth for a pixel.
    """
    # Number of match point correspondences
    n_points = matched_points.shape[0]
    # Estimate R, T matrices using bundle adjustment
    estimated_parameters = bundle_adjustment(n_frames, matched_points)

    # Separate out the rotation angles, translation matrices from estimated parameters
    R_angles = estimated_parameters.x[:n_frames * 6].reshape(n_frames,6)[:,:3]
    T_matrices = estimated_parameters.x[:n_frames * 6].reshape(n_frames,6)[:,3:]

    # Get the depths of the candidate planes
    depths = 1/estimated_parameters.x[n_frames*6:]
    candidate_depths = np.linspace(min(depths), max(depths), 10)

    # Height and width of the images
    h, w = images_array[0].shape
    variance_cost = np.zeros((h, w, 10))

    for j in range(10):
        warped_images = np.zeros((h, w, n_frames))
        for i in range(n_frames):
            r = R_angles[i]
            # Prepare the rotation matrix from the estimated rotation angles for each frame
            R = np.array([[1, -r[2], r[1]], [r[2], 1, -r[0]], [-r[1], r[0], 1]])
            middle_term = R - np.dot(T_matrices[i].reshape(-1,1), np.array([[0,0,-1]]) ) / candidate_depths[j]
            # Plane induced homography matrix
            H = np.dot(np.dot(K, middle_term), np.linalg.inv(K))
            # Warp each image to the frame of the reference image
            warped_images[:, :, i] = cv2.warpPerspective(images_array[i], np.linalg.inv(H) ,(w,h))
            
        # Calculate the variance in the frames axis
        variance_cost[:, :, j] = np.var(warped_images,axis=2)

    # Choose the candidate depth that minimizes variance
    depth_map = 1/candidate_depths[np.argmin(variance_cost,axis=2)]

    return depth_map
```

### Plane Sweep Stereo using 5 frames

In the depth maps, the brighter the pixel, the closer it is to the screen (reference).

```python
n_frames = 5
depth_map = plane_sweep_stereo(n_frames, matched_points, images_array)

plt.figure(figsize=(12, 14))
plt.subplot(1, 2, 1)
plt.imshow(images_array[0], cmap="gray")
plt.title("Reference image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(depth_map, cmap="gray")
plt.axis("off")
plt.title("Depth map (5 frames)")
plt.show()
```
