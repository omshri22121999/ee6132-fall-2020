## Part 1 - Structure From Motion

**Reading the matching image points and camera intrinsic matrix**

```python
# Loading matching points
mp = io.loadmat("./matchedPoints.mat")["matchedPoints"]

# Loading Camera intrinsic matrices
ks = io.loadmat("./IntrinsicsMCV.mat")["IntrinsicMatrix"]
```

**We extract the focal lengths and principal points from the intrinsic camera matrix**

```python
# Setting camera focus
fx, fy = ks[0][0], ks[1][1]

# Setting camera translation
cx, cy = ks[0][-1], ks[1][-1]

print("Camera Characteristics : \n")
print("fx :", fx, " fy :", fy, "\n", "\bcx :", cx, "cy :", cy)
```

**The function `norm_img_pts` converts the image coordinates to camera coordinates**

```python
def norm_img_pts(mp, cx, cy, fx, fy):

    # Copy the original match points to new variable for manipulation
    mp_norm = mp.copy()

    # Convert the image coordinates to camera coordinates
    mp_norm[:, 0, :] = (mp_norm[:, 0, :] - cx) / fx
    mp_norm[:, 1, :] = (mp_norm[:, 1, :] - cy) / fy

    return mp_norm
```

```python
# Procuring the camera coordinates from image coordinates
mp_c = norm_img_pts(mp, cx, cy, fx, fy)
```

**The function `projection_error` computes the projection error for the original matched points and the camera points using the fundamental matrix. We try to minimize the error to get the parameters of the fundamental matrix and depth of the matching points**

```python
def projection_error(params, match_points):

    # Extracting the depth params and converting to required shape for vectorized computation
    w = params[: match_points.shape[2]].reshape((1, match_points.shape[2]))
    w_st = np.stack([w for i in range(match_points.shape[0])], axis=0)

    # Stack the first image camera coordinates which is required for vectorized computation
    mp_0_st = np.stack([match_points[0] for i in range(match_points.shape[0])], axis=0)

    # Extracting the rotation matrix params and converting to required shape for vectorized computation
    theta = params[
        match_points.shape[2] : match_points.shape[2] + 3 * match_points.shape[0]
    ].reshape((match_points.shape[0], 3, 1))

    # Creating the rotation matrix from the params
    rot = np.ones((match_points.shape[0], 3, 3))

    rot[:, 1, 0] = theta[:, 2, 0]
    rot[:, 0, 1] = -theta[:, 2, 0]
    rot[:, 0, 2] = theta[:, 1, 0]
    rot[:, 2, 0] = -theta[:, 1, 0]
    rot[:, 2, 1] = theta[:, 0, 0]
    rot[:, 1, 2] = -theta[:, 0, 0]

    # Extracting the translation params and converting to required shape for vectorized computation
    tr = params[match_points.shape[2] + 3 * match_points.shape[0] :].reshape(
        (match_points.shape[0], 3, 1)
    )

    # Computing the projection using the above params
    proj_match_points = (
        np.matmul(
            rot,
            np.append(
                mp_0_st,
                np.ones((match_points.shape[0], 1, match_points.shape[2])),
                axis=1,
            )
            * w_st,
        )
        - tr
    )

    # Converting the projection from homogeneous 3D to 2D form
    proj_match_points = np.stack(
        [
            proj_match_points[:, 0, :] / proj_match_points[:, 2, :],
            proj_match_points[:, 1, :] / proj_match_points[:, 2, :],
        ],
        axis=1,
    )

    # Returning the absolute difference between the projection and original matched points
    return (match_points - proj_match_points).ravel()
```

```python
def projection_function(params, match_points):
    w = params[: match_points.shape[2]].reshape((1, match_points.shape[2]))

    mp_0_3d = w * np.r_[match_points[0], np.ones((1, match_points[0].shape[1]))]

    cam_params = params[match_points.shape[2] :].reshape(match_points.shape[0], 6)

    error_x = []
    error_y = []

    for i in range((match_points.shape[0])):
        theta = cam_params[i, :3]
        rot_mat = np.array(
            [
                [1, -theta[2], theta[1]],
                [theta[2], 1, -theta[0]],
                [-theta[1], theta[0], 1],
            ]
        )
        tr_mat = cam_params[i, 3:].reshape(3, 1)

        proj_3d = np.matmul(rot_mat, mp_0_3d) + tr_mat

        proj_2d = np.array(
            [proj_3d[0, :] / proj_3d[2, :], proj_3d[1, :] / proj_3d[2, :]]
        )

        error_x.append((proj_2d - match_points[i])[0])
        error_y.append((proj_2d - match_points[i])[1])

    return np.array(error_x + error_y).ravel()
```

```python
def ProjectionFunction(threeDpoints, camera_parameters, n_points, n_frames):

    # Given the 3D points and all the camera parameters for each frame, this function projects the 3D points with respect to each frame
    projected_points = np.zeros((n_frames, n_points, 3))

    for i in range(n_frames):
        # i th frame, the first

        theta = camera_parameters[i, :3]
        rot_mat = np.array(
            [
                [1, -theta[2], theta[1]],
                [theta[2], 1, -theta[0]],
                [-theta[1], theta[0], 1],
            ]
        )
        rotated_threeDpoints = np.matmul(threeDpoints, rot_mat)

        # To the rotated coordinates, add the translation vector (last 3 values)

        T_matrix_i = camera_parameters[i, 3:]
        projected_points[i] = rotated_threeDpoints + T_matrix_i

    # Converting 2D homogeneous to heterogeneous coordinates
    projected_x = projected_points[:, :, 0] / projected_points[:, :, 2]
    projected_y = projected_points[:, :, 1] / projected_points[:, :, 2]

    return projected_x.T, projected_y.T
```

```python
def reprojection_error_residuals(params, p):
    n_points = p.shape[0]

    n_frames = p.shape[2]

    camera_parameters = params[: n_frames * 6].reshape((n_frames, 6))
    W = params[n_frames * 6 :].reshape((n_points, 1))

    threeDpoints = np.c_[p[:, 0, 0], p[:, 1, 0], np.ones((n_points, 1))] * W

    projected_x, projected_y = ProjectionFunction(
        threeDpoints, camera_parameters, n_points, n_frames
    )

    errors_in_x = projected_x - p[:, 0, :n_frames]
    errors_in_y = projected_y - p[:, 1, :n_frames]

    return np.hstack((errors_in_x, errors_in_y)).ravel()
```

**The function `optimize_projection` finds the best fundamental matrix parameters and the best depth values for the reference image (here I use the first image). We do somewhat optimal initialization of params by setting all the depth parameters as 1 and rotation and translation parameters as 0**

```python
def optimize_projection(n_frames, mp):
    # Initialising the depth parameters as 1's
    x0 = np.ones((6 * n_frames) + mp.shape[0])

    # Initialising the rotation and translation parameters as 0's
    x0[mp.shape[0] :] = 0

    # Running the least squares optimization algorithm
    out = op.least_squares(
        projection_function,
        x0,
        args=([mp[:, :, :n_frames]]),
        x_scale="jac",
        verbose=2,
        max_nfev=20,
    )

    # Extracting the minimized parameters
    params = out["x"]

    # Extracting the depth paramters
    w = params[: mp.shape[0]].reshape(1, -1)

    # Extracting the rotation parameters
    theta = params[mp.shape[0] : mp.shape[0] + 3 * mp.shape[2]].reshape(
        (mp.shape[2], 3, 1)
    )

    #     rot = np.ones((mp.shape[0], 3, 3))

    #     rot[:, 1, 0] = theta[:, 2, 0]
    #     rot[:, 0, 1] = -theta[:, 2, 0]
    #     rot[:, 0, 2] = theta[:, 1, 0]
    #     rot[:, 2, 0] = -theta[:, 1, 0]
    #     rot[:, 2, 1] = theta[:, 0, 0]
    #     rot[:, 1, 2] = -theta[:, 0, 0]

    #     # Extracting the translation parameters
    #     tr = params[mp.shape[2] + 3 * mp.shape[0] :].reshape((mp.shape[0], 3, 1))

    #     return w, rot, tr

    cam_params = params[mp.shape[0] :].reshape(mp.shape[2], 6)

    rot_mats = []
    tr_mats = []

    for i in range((mp.shape[2])):
        theta = cam_params[i, :3]
        rot_mats.append(
            np.array(
                [
                    [1, -theta[2], theta[1]],
                    [theta[2], 1, -theta[0]],
                    [-theta[1], theta[0], 1],
                ]
            )
        )
        tr_mats.append(cam_params[i, 3:].reshape(3, 1))

    rot_mats = np.array(rot_mats)
    tr_mats = np.array(tr_mats)
    return w, rot_mats, tr_mats
```

```python
# Extracting R,T and W using all the matching point frames

params = optimize_projection(mp_c.T.shape[0], mp_c.T)
```

```python
w = params[0]

rot = params[1]

tr = params[2]
```

```python
print(rot)
```

```python
# Reading images

images = []
img_locs = glob.glob("./sequence/*")

for img_loc in img_locs:
    images.append(cv2.cvtColor(cv2.imread(img_loc), cv2.COLOR_BGR2GRAY))
```

```python
# Setting the number of candidate depth planes
n_d = 10

# Creating the candidate depth planes
d_space = np.linspace(np.min(w), np.max(w), n_d)
```

**The function `generate_depth_map` creates multiple homographies of above sequence of images and stacks them**

```python
def generate_depth_map(n_frames):

    # Creating the tensor to hold the result of homographic transformations
    img_space = np.zeros((images[0].shape[0], images[0].shape[1], n_d, n_frames))

    # Iterating through each image and each depth candidate
    for i in range(n_frames):
        for j in range(n_d):
            # Computing the homography matrix
            hm_mat = np.matmul(
                np.matmul(
                    ks, (rot[i] - np.matmul(np.array([0, 0, -1]), tr[i]) * d_space[j])
                ),
                np.linalg.inv(ks),
            )

            # Applying the homography transform
            out_img = cv2.warpPerspective(images[i], np.linalg.inv(hm_mat), None)
            img_space[:, :, j, i] = out_img

    # Creating the variance space
    img_var = np.var(img_space, axis=3)

    # Finding the index of the minimum variance depth for each pixel
    d_min = np.argmin(img_var, axis=2)

    # Creating the depth map
    depth_map = d_space[d_min]

    return depth_map
```

### Depth Map using 25 frames

```python
depth_map_25 = generate_depth_map(25)
```

```python
plt.figure(figsize=(10, 10))
plt.title("25 Images")
plt.imshow(
    np.c_[
        (
            (depth_map_25 - depth_map_25.min())
            * 255
            / (depth_map_25.max() - depth_map_25.min())
        ),
        images[0],
    ],
    cmap="gray",
)
plt.axis("off")
plt.show()
```
